```python
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

### **Loading the Image**
<hr/>

```python
image= cv2.imread(r"C:\Users\Jocelyne\Downloads\IMG_20230124_100524.jpg")
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

### Converting the Image to Greyscale
<hr/>

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_image,cmap='gray')
plt.axis("off")
plt.title("Grayscale Image")
plt.show()
```

### Invert the Grayscale Image
<hr/>

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.figure(figsize=(8,8))
plt.imshow(inverted_image,cmap='gray')
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

### Smoothing the Image
<hr/>

```python
smoothened_image = cv2.GaussianBlur(inverted_image,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(smoothened_image,cmap='gray')
plt.axis("off")
plt.title("Smoothened Image")
plt.show()
```

### Converting the Image to Pencil Sketch
<hr/>

```python
pencil_sketch_image = cv2.divide(gray_image, 255-smoothened_image, scale= 255)
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch_image,cmap='gray')
plt.axis("off")
plt.title("Pencil Sketch Image")
plt.show()
```
